# Prepare the Data

```python
import pandas as pd
```

```python
df = pd.read_excel('supermarkt_sales.xlsx',skiprows=3)
df_copy = df.copy()
```

# First Column

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
# df.groupby(df['Product line']).sum()
df.groupby(by=['Product line']).sum()
```

```python
df.groupby(df['Product line']).sum()['Total']
```

```python
df.groupby(df['Product line']).sum()[['Total']].sort_values('Total')
```

# Second Column

```python
df_copy = df_copy.drop('Unnamed: 0', axis=1)
```

```python
# The time column is still of type object
df_copy.info()
```

```python
#The data below cannot be converted into hours, because the format has not been changed to datetime
df_copy['hour'] = df['Time'].dt.hour
```

```python
#change the object data type to datetime with format='%H:%M:%S' then .dt.hour
df_copy['hour'] = pd.to_datetime(df_copy['Time'], format='%H:%M:%S').dt.hour
df_copy
```
